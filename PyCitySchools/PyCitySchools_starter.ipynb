```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(50)
```

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = len(school_data_complete["Student ID"])
total_budget = sum(school_data_complete["budget"].unique())
avg_math = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()
percent_pass_math = school_data_complete["math_score"].astype("float").loc[(school_data_complete["math_score"])>=70].count()/total_students*100
percent_pass_reading = school_data_complete["reading_score"].astype("float").loc[(school_data_complete["reading_score"])>=70].count()/total_students*100
percent_overall = len(school_data_complete[["reading_score","math_score"]].astype("float").loc[(school_data_complete["math_score"]>= 70) & (school_data_complete["reading_score"]>= 70)])/total_students*100
df = pd.DataFrame({"Total Schools": [total_schools],
                  "Total Students": total_students,
                  "Total Budget":total_budget,
                  "Average Math Score": avg_math,
                  "Average Reading Score": avg_reading,
                  "% Passing Math": percent_pass_math,
                  "% Passing Reading": percent_pass_reading,
                 "% Overall Passing": percent_overall})
df["Total Budget"] = df["Total Budget"].map("${:,.2f}".format)
df["% Passing Math"]= df["% Passing Math"].map("{:.2f}%".format)
df["% Passing Reading"]= df["% Passing Reading"].map("{:.2f}%".format)
df["% Overall Passing"]= df["% Overall Passing"].map("{:.2f}%".format)
df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
groupby= school_data_complete.groupby("school_name")
school_type = groupby["type"].unique().rename("School Type")
students = groupby["Student ID"].count().rename("Total Students")
budget = groupby["budget"].unique().astype("float").rename("Total School Budget")
pp_budget = (budget/students).rename("Per Student Budget")
ag_math = groupby["math_score"].mean().rename("Average Math Score")
ag_reading = groupby["reading_score"].mean().rename("Average Reading Score")
pass_math = school_data_complete[["math_score","school_name"]].loc[(school_data_complete["math_score"])>=70].groupby("school_name").count().astype("float")
percent_math = (pass_math["math_score"]/students*100).rename("% Passing Math")
pass_reading = school_data_complete[["reading_score","school_name"]].loc[(school_data_complete["reading_score"])>=70].groupby("school_name").count().astype("float")
percent_reading = (pass_reading["reading_score"]/students*100).rename("% Passing Reading")
pass_both = school_data_complete.loc[(school_data_complete["math_score"]>= 70) & (school_data_complete["reading_score"]>= 70),["school_name","reading_score"]].groupby("school_name").count()
percent_pass = (pass_both["reading_score"]/students*100).rename("% Overall Passing")
df2 = school_type.to_frame().join(students).join(budget).join(pp_budget).join(ag_math).join(ag_reading).join(percent_math).join(percent_reading).join(percent_pass).reset_index()
df2["Per Student Budget"] = df2["Per Student Budget"].map("${:,.2f}".format)
df2["Total School Budget"] = df2["Total School Budget"].map("${:,.2f}".format)
df2["% Passing Math"]= df2["% Passing Math"].map("{:.2f}%".format)
df2["% Passing Reading"]= df2["% Passing Reading"].map("{:.2f}%".format)
df2["% Overall Passing"]= df2["% Overall Passing"].map("{:.2f}%".format)
df2
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0,585,630,645,676]
labels = ["<$584","$585-629","$630-644","$645-675"]
school_data_complete["Spending Ranges (Per Student)"]= pd.cut((school_data_complete["budget"]/school_data_complete["size"]),bins,labels=labels,include_lowest = True,right=False)
grouped = school_data_complete.groupby("Spending Ranges (Per Student)")
avg_math = grouped["math_score"].mean().rename("Average Math Score")
avg_reading = grouped["reading_score"].mean().rename("Average Reading Score")
percent_math = school_data_complete.loc[school_data_complete["math_score"]>= 70].groupby("Spending Ranges (Per Student)").count()
math=(percent_math["math_score"]/grouped["math_score"].count()*100).rename("% Passing Math")
percent_reading = school_data_complete.loc[school_data_complete["reading_score"]>= 70].groupby("Spending Ranges (Per Student)").count()
reading=(percent_reading["reading_score"]/grouped["reading_score"].count()*100).rename("% Passing Reading")
percent_overall = school_data_complete.loc[(school_data_complete["math_score"]>= 70) & (school_data_complete["reading_score"]>= 70)].groupby("Spending Ranges (Per Student)").count()
overall = (percent_overall["reading_score"]/grouped["reading_score"].count()*100).rename("% Overall Passing")
#overall = ((math+reading)/2).rename("% Overall Passing")
df9= avg_math.to_frame().join(avg_reading).join(math).join(reading).join(overall).reset_index()
df9["% Passing Math"]= df9["% Passing Math"].map("{:.2f}%".format)
df9["% Passing Reading"]= df9["% Passing Reading"].map("{:.2f}%".format)
df9["% Overall Passing"]= df9["% Overall Passing"].map("{:.2f}%".format)
df9["Average Math Score"]= df9["Average Math Score"].map("{:.2f}".format)
df9["Average Reading Score"]= df9["Average Reading Score"].map("{:.2f}".format)
df9= df9.set_index("Spending Ranges (Per Student)")
df9
```

```python
bins = [0,1000,2000,5000]
labels = ["Small(<1000)","Medium(1000-2000)","Large(2000-5000)"]
school_data_complete["School Size"]= pd.cut(school_data_complete["size"],bins,labels=labels,include_lowest = True)
grouped = school_data_complete.groupby("School Size")
avg_math = grouped["math_score"].mean().rename("Average Math Score")
avg_reading = grouped["reading_score"].mean().rename("Average Reading Score")
percent_math = school_data_complete.loc[school_data_complete["math_score"]>= 70].groupby("School Size").count()
math=(percent_math["math_score"]/grouped["math_score"].count()*100).rename("% Passing Math")
percent_reading = school_data_complete.loc[school_data_complete["reading_score"]>= 70].groupby("School Size").count()
reading=(percent_reading["reading_score"]/grouped["reading_score"].count()*100).rename("% Passing Reading")
percent_overall = school_data_complete.loc[(school_data_complete["math_score"]>= 70) & (school_data_complete["reading_score"]>= 70)].groupby("School Size").count()
overall = (percent_overall["reading_score"]/grouped["reading_score"].count()*100).rename("% Overall Passing")
#overall = ((math+reading)/2).rename("% Overall Passing")
df9= avg_math.to_frame().join(avg_reading).join(math).join(reading).join(overall).reset_index()
df9["% Passing Math"]= df9["% Passing Math"].map("{:.2f}%".format)
df9["% Passing Reading"]= df9["% Passing Reading"].map("{:.2f}%".format)
df9["% Overall Passing"]= df9["% Overall Passing"].map("{:.2f}%".format)
df9["Average Math Score"]= df9["Average Math Score"].map("{:.2f}".format)
df9["Average Reading Score"]= df9["Average Reading Score"].map("{:.2f}".format)
df9= df9.set_index("School Size")
df9
```
